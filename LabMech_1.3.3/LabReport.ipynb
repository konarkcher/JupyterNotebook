### Основные теоретические сведения

Характер движения газа (или жидкости) в трубке определяется безразмерными числом Рейнольдса:

$$
Re=\dfrac{\upsilon r\rho}\eta, \tag1
$$

где $\upsilon$ — скорость потока, $r$ — радиус трубки, $\rho$ — плотность
движущейся среды, $\eta$ — её вязкость.

При ламинарном течении объём газа $V$, протекающий за время $t$ по трубе длиной $l$, определяется формулой
Пуазейля: 

$$
Q_v=\dfrac{\pi r^4}{8l\eta}(P_1-P_2) \tag2
$$

Величину $Q$ обычно называют расходом.

Условия справедливости формулы (1):

1. c достаточным запасом выполняется $Re<1000$;
2. не происходит существенного изменения удельного объёма вещества.

Для жидкости последнее выполняется практически всегда, а для газа - если перепад давления мал по сравнению с самим
давлением.

В нашем случае давление равно атмосферному (10<sup>3</sup> см вод. ст.), а перепад состаляет не более 10 см,
т.е. менее 1% от атмосферного.

Формула (1) выводится для участков трубы, на которых закон распределения скоростей газа по сечению не меняется
при движении вдоль потока.

![Профиль скоростей](gas_stream.png)

Характерное распределение скоростей формируется на некотором расстоянии $a$ от входа в трубку:

$$
    a\approx0,2\cdot Re. \tag3
$$
